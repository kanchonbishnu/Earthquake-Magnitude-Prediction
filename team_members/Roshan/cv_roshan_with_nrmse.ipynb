```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_val_score
import numpy as np
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import ElasticNet
from catboost import CatBoostRegressor
from sklearn.neural_network import MLPRegressor


```

```python
# Initialize and train models with the preprocessor
models = {
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "Support Vector Regression": SVR(),
    "XGBoost": XGBRegressor(),
    "LightGBM": LGBMRegressor(),
    "Catboost": CatBoostRegressor(),
    "ElasticNet": ElasticNet(),
    "BayesianRidge": BayesianRidge(),
    "MLP": MLPRegressor(max_iter=500)
}
```

```python
results = {}
```

```python
def evaluate_model_cv(model, X_train, y_train, cv=5):
    mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
    mse_scores = cross_val_score(model, X_train, y_train, scoring=mse_scorer, cv=5)
    mean_mse = -np.mean(mse_scores)
    return mean_mse
```

```python
for model_name, model in models.items():
    pipeline = Pipeline(steps=[('regressor', model)])
    mean_mse = evaluate_model_cv(pipeline, X_train, y_train, model)
    results[model_name] = {
        'mean_mse': mean_mse,

    }
```

```python
sorted_results = sorted(results.items(), key=lambda x: x[1]['mean_mse'])
```

```python
sorted_results
```

```python
results_df = pd.DataFrame([{"Model": model_name,"Mean Squared Error": metrics['mean_mse']}for model_name, metrics in results.items()])
results_df = results_df.sort_values(by="Mean Squared Error")
```

```python
results_df['Root Mean Squared Error'] = np.sqrt(results_df['Mean Squared Error'])
results_df.reset_index(drop=True,inplace=True)
print(results_df)
```

```python
from sklearn.ensemble import StackingRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
stacking_model = StackingRegressor(
    estimators=[
        ('catboost', CatBoostRegressor(verbose=0)),
        ('lightgbm', LGBMRegressor()),
        ('rf', RandomForestRegressor())
    ]
)
```

```python
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
```

```python
pipeline = Pipeline(steps=[('stacking', stacking_model)])
pipeline.fit(X_train, y_train)
```

```python

# Perform cross-validation to get the mean MSE
mse_scores = cross_val_score(pipeline, X_train, y_train, scoring=mse_scorer, cv=5)
mean_mse = -np.mean(mse_scores)  # Convert back to positive MSE
print(f"Stacking Regressor - Mean Cross-Validated MSE: {mean_mse:.4f}")
```

```python
results_df
```

```python
results_df.loc[12] = ["Ensemble Learning", mean_mse, np.sqrt(mean_mse)]
results_df = results_df.sort_values(by='Mean Squared Error', ascending=True).reset_index(drop=True)
```

```python
def calculate_nrmse(mse, targets):

    rmse = np.sqrt(mse)
    # Calculate range of target values
    target_range = targets.max() - targets.min()

    # Calculate NRMSE as a percentage
    nrmse = (rmse / target_range) * 100

    return nrmse

# Calculate NRMSE using magnitude values from earthquake_df
results_df['NRMSE (%)'] = results_df['Mean Squared Error'].apply(
    lambda x: calculate_nrmse(x, earthquake_df['mag'])
)

# Round to 2 decimal places
results_df['NRMSE (%)'] = results_df['NRMSE (%)'].round(2)

# Display the updated DataFrame
print(results_df)
```
